# Marginal Probabilities(Single column)

1.What is probability of getting rain tomorrow

```python
sum(df['Outlook']=='Rain')/len(df)
```

```python
df['Outlook'].value_counts(normalize=True).loc['Rain']
```

2.What is the probability of playing cricket tommorow

```python
sum(df['Play']=='Yes')/len(df)
```

```python
df['Play'].value_counts(normalize=True).loc['Yes']
```

3.What is the probability that temperature is not cool tommorow

```python
sum(df['Temperature']!='Cool')/len(df)
```

```python
1 - df['Temperature'].value_counts(normalize=True).loc['Cool']
```

# Joint Probabilities

1) What is the probability of playing cricket and temperature is hot?

```python
sum((df['Play']=='Yes')&(df['Temperature']=='Hot'))/len(df)
```

```python
pd.crosstab(df['Play'],df['Temperature'],normalize=True).loc['Yes']['Hot']
```

2.What is the probability of humidity is normal and temperature is cool

```python
pd.crosstab(df['Humidity'],df['Temperature'],normalize=True).loc['Normal']['Cool']
```

# Conditional Probability

1. What is the probability of playing cricket if it is raining

```python
pd.crosstab(df['Play'],df['Outlook'],normalize='columns').loc['Yes']['Rain']
```

2. What is the probability of not playing cricket when humidity is high

```python
pd.crosstab(df['Play'],df['Humidity'],normalize='columns').loc['No']['High']   #in normalize we should keep always the event which is happened already
```
